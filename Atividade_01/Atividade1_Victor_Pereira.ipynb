Nesse trecho de código, é feita a importação das bibliotecas: numpy, linalg e pyplot, quer  serão utilizadas no nosso código. Também é feita uma abreviação dos seus nomes

````python
import numpy as np
import numpy.linalg as lg
import matplotlib.pyplot as plt
````

<br>
Nesse trecho de código, é criado o vetor que representa o eixo das Abscissas, o primeiro ponto é o -15 e último é o 15. O último parâmetro indica que temos 3000 pontos entre o ponto inicial e o ponto final.

````python
# Dimensão x
x = np.linspace(-15, 15, 3000)
````

<br>
Nesse trecho de código, estamos definindo nosso eixo y, ou seja, o eixo das ordenadas. Estamos atribuindo a variável y, o valor de $x^3$

````python
# Função f(x)
y = x**3
````

<br>
Nesse trecho de código, estamos calculando o valor da derivada de y e armazenando o resultando na variável yl. O valor no caso seria o resultado da expressão: $3x^2$

````python
# Derivada exata
yl = 3*x**2
````

<br>
Neste momento da execução, é calculado o delta do eixo x, subtraindo de um ponto o seu antecessor, conforme a seguinte fórmula:

$x - (x-1)$

````python
deltaX = x[1:] - x[:-1]
````

<br>
De maneira semelhante, é calculado o delta para o eixo y: 

$y - (y - 1)$

````python
deltaY = y[1:] - y[:-1]
````

<br>
Agora, calculamos o coeficiente angular com a seguinte fórmula: 

$\frac{\Delta\\y}{\Delta\\x}$

E armazenamos na variável yh

````python
yh = deltaY / deltaX
````

<br>
A variável error armazena o erro relativo, com base na seguinte fórmula:

$\frac{\delta\\y - \alpha}{\Delta\\y}$

````python
# Cálculo do erro relativo na norma L2
error = (lg.norm(yl) - lg.norm(yh)) / lg.norm(yl)
````

<br>

No trecho de código abaixo, iniciamos a plotagem dos gráficos, segue a definição das funções utilizadas:

Fechamos todas as figuras plotadas anteriormente

```python
plt.close('all')
```


A função plot cria um gráfico de uma reta, onde passamos as coordenadas de x e y, passamos também o label com o seguinte texto: $f(x)$

```python
plt.plot(x, y, label='f(x)')
```

A função plot cria um gráfico de uma parábola, onde passamos as coordenadas de x e y, passamos também o label com o seguinte texto: Derivada aproximada

```python
plt.plot(x[:-1], yh, label="Derivada Aproximada")
```

A função plot cria um gráfico de uma parábola, onde passamos as coordenadas de x e y, passamos também o label com o seguinte texto: Derivada exata


```python
plt.plot(x, yl, label="Derivada Exata", linestyle='--')
```

Aqui adicionamos uma legenda ao gráfico

```python
plt.legend()
```

Aqui adicionamos os labels para as coordenadas x e y, respectivamente

```python
plt.xlabel('x')
plt.ylabel('y')
```

Aqui adicionamos o título do gráfico

```python
plt.title('Função e suas Derivadas\nErro Relativo L2: {:.4e}'.format(error))
```
E, finalmente a função show plota o gráfico na tela

```python
plt.grid(True)
plt.show()
```

````python
# Plots
plt.close('all')

plt.figure(1)
plt.plot(x, y, label='f(x)')
plt.plot(x[:-1], yh, label="Derivada Aproximada")
plt.plot(x, yl, label="Derivada Exata", linestyle='--')

plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função e suas Derivadas\nErro Relativo L2: {:.4e}'.format(error))
plt.grid(True)
plt.show()
````
